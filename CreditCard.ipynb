```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_data = pd.read_csv('/content/drive/MyDrive/creditcard.csv/creditcard.csv')
```

```python
#dataset information
credit_data.info()
```

```python
#distribution of legit transactions & fradulent transactions
credit_data['Class'].value_counts()
```

Dataset is highly imbalanced


0 --> Normal Transaction

1 --> Fradulent Transaction





```python
#Separate data for analysis
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare values for both transactions
credit_data.groupby('Class').mean()
```

UnderSampling

Build a sample dataset containing similar distribution of normal transactions & fradulent transactions

No. of Fraudulent Transactions = 492

```python
legit_sample = legit.sample(n=492)
```

Concatenate 2 dataframes

```python
#axis =1 (column) , axis=0 (row)
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Since mean values are simmilar so we have got good sample and the nature of dataset is not changed.

Splitiing data into Features & Targets

```python
X = new_dataset.drop(columns = 'Class', axis=1)
Y = new_dataset['Class']
```

```python
# X contains features
print(X)
```

```python
#Y contains the labels
print(Y)
```

Split data into Train and Test data

```python
#Train data = 80% of data  & Test data = 20% of data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2 )
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the model

Using Logistic Regression

which is used generally for binary classification problems

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression model with Train data
model.fit(X_train, Y_train)
```

Model Evaluation based on Accuracy score

```python
#Accuracy Score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Train data :', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test data: ', test_data_accuracy)
```

Since the accuracy of Train data and Test data are similar so the model is not overfitted(over-tested) or underfitted (over-trained)
